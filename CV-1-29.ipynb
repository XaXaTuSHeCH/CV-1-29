```python
from skimage import data
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
astro = data.astronaut()
astro_mirror = astro[:,::-1]
result = cv2.hconcat([astro, astro_mirror])
result = cv2.line(result, (512,0), (512,512), (255,255,255), 5)
plt.imshow(result)
plt.axis('off')
plt.show()
```
